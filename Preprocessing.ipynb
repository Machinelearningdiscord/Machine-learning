Read salaries.csv


```
import pandas as pd
import numpy as np
salaries_csv="https://raw.githubusercontent.com/Machinelearningdiscord/Machine-learning/master/Salaries.csv"
df=pd.read_csv(salaries_csv,low_memory=False)
```

Exploring dataframe

```
pd.set_option('max_columns', None)

df.TotalPayBenefits[1:50]='?'
print(df.head())
print(df.info())
```

```
#checking for unneccary values in TotalPayBenefits
df.TotalPayBenefits.values

#df[df.TotalPayBenefits.values=='?']
#df.info()
```

```
df.TotalPayBenefits[df.TotalPayBenefits=='?']=np.nan
print(df.TotalPayBenefits.values)
print(df.TotalPayBenefits.isnull().sum())
```

convert nan to mean of the feature vector


```
df.TotalPayBenefits=df.TotalPayBenefits.fillna(df.TotalPayBenefits.mean())
print([df.TotalPayBenefits.isnull().sum()])
df.TotalPayBenefits
```

Dealing with Categorical/ordinal Variables

How much does   "JOSEPH DRISCOLL" make (Inlcuding benefits)


```
np.random.seed(3)
x=np.random.randint(120,180,10)
y=['Naveen','Dwithi','Danish','Pooja','Akshay','Lisa','Priya','Fareen','Nisha','Ankita']
z=['M','F','M','F','M','F','F','F','F','F',]
#print(sch)
sch=pd.DataFrame({'height':x,'Name':y,'Gender':z})
dummy=pd.get_dummies(sch.Gender)
sch
```

drop Male frm Dummy


```
dummy=dummy.drop('M',axis=1)
sch=sch.drop('Gender',axis=1)
sch=pd.concat([sch,dummy],axis=1)
sch
```

Normalization


```
np.random.seed(3)
PE_marks=pd.DataFrame({'PE_marks':np.random.randint(500,1000,10)})
sch=pd.concat([sch,PE_marks],axis=1)
sch
```

```
sch.height = (sch.height - sch.height.mean())/sch.height.std()
sch.PE_marks = (sch.PE_marks - sch.PE_marks.mean())/sch.PE_marks.std()
sch
```

```
import matplotlib.pyplot as plt
plt.scatter(sch.height.values,sch.PE_marks.values)
plt.xlabel('Height')
plt.ylabel('PE_marks')
plt.show()
```

What are top 5 common jobs


```
import matplotlib.pyplot as plt
plt.scatter(sch.F.values,sch.PE_marks.values)
plt.xlabel('Gender')
plt.ylabel('PE_marks')
plt.show()
```

Correlation 


```
correlation_matrix = np.corrcoef(sch.F.values,sch.PE_marks.values)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```

```
correlation_matrix = np.corrcoef(sch.height.values,sch.PE_marks.values)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(r_squared)
```
